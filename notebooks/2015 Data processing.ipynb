# Translate Categorical Variables

This applies only to variables with more than 2 categories (non binary)

So here i just used 2 approaches for demonstration. 
    1. Creating function then use .apply
    2. use .map
    
Both will work fine but in case you would need to recreate the cleaning and function better to use approach #1


```python
#BMI Category Function

# defining function to BMI
def bmi_calc(a):
  
    if a<18.5:
        return "1. Underweight"
  
    elif a<25:
        return "2. Normal"
    elif a<30:
        return "3. Overweight"
    elif a<35:
        return "4. Obese 1"
    elif a<40:
        return "5. Obese 2"
    else:
        return "Obese 3"
```

```python
#Diabetes Category Function

# defining function to Diabetis History
def diabetes_calc(a):
  
    if a==0:
        return "No or During Pregnancy"
  
    elif a<25:
        return "Pre Diabetes/Borderline"
    else:
        return "Existing Diabetes"
```

```python
data['bmi_category']=data['BMI'].apply(bmi_calc)
```

```python
data['diab_category']=data['Diabetes'].apply(diabetes_calc)
```

```python
data['age_grp']=data['Age'].map({
    1:'18-24',
    2:'25-29',
    3:'30-34',
    4:'35-39',
    5:'40-44',
    6:'45-49',
    7:'50-54',
    8:'55-59',
    9:'60-64',
    10:'65-69',
    11:'70-74',
    12:'75-79',
    13:'80 and up'
})
```

```python
data['educ_grp']=data['Education'].map({
    1:'Never attended school / Only Kindergarten',
    2:'Grades 1 - 8 (elementary)',
    3:'Grades 9-11 (Some High School)',
    4:'Grade 12 or GED (High School Graduate)',
    5:'College 1-3 (Some College/Technical School)',
    6:'College Graduate'
})
```

```python
data['inc_grp']=data['Income'].map({
    1:'Less than 10k USD',
    2:'10k-15k USD',
    3:'15k-20k USD',
    4:'20k-25k USD',
    5:'25k-35k USD',
    6:'35k-50k USD',
    7:'50k-75k USD',
    8:'more than 75k USD'
})
```

# One Hot Encode

Create dummies for categorical variables, subsequently drop columns wherein dummies are created.


```python
to_drop = ['BMI','Age','Diabetes','Education','Income']
for col in tqdm(data.columns):
    if data[col].dtype == 'O':
        dummies = pd.get_dummies(data[col], prefix=col, drop_first=False)
        heart_data = pd.concat([data, dummies], axis=1)
        to_drop.append(col)
heart_data = heart_data.drop(to_drop, axis=1)
heart_data.head()
```

# Check for Outliers

Checking for outliers only applies to numerical/continuous variables. As you will see here I only applied to certain variables.


```python
numerical_cols = ['MentHlth', 'BMI', 'PhysHlth', 'HeartDiseaseorAttack']
heart_quant = data[numerical_cols]
sns.pairplot(heart_quant, hue='HeartDiseaseorAttack')
plt.show()
```

```python
heart_quant2 = heart_quant[['MentHlth', 'BMI', 'PhysHlth']]

# initialize baseline
iso=IsolationForest(random_state=143)
iso.fit(heart_quant2)

# prediction of outliers is based on contamination level
y_pred_IF = iso.fit_predict(heart_quant2)
IF_scores = iso.score_samples(heart_quant2)
```

```python
sns.displot(IF_scores, bins=60, color="#F25278")
plt.axvline(-0.75, linestyle='--', color='r')
plt.show()
```

Based on the Plot above no need for exclusion of outlier. No apparent outlier detected.
